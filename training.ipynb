# **Next word prediction LSTM model**

```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense
```

### Reading text file
- when reading file line breaks become newline characters('\n')

```python
with open("data/Alice's Adventures in Wonderland by Lewis Carroll.txt", 'r', encoding='utf-8') as myfile:
    alices_dventures_txt = myfile.read()
```

```python
alices_dventures_txt
```

### Tokenizing

```python
my_tokenizer = Tokenizer()
my_tokenizer.fit_on_texts([alices_dventures_txt])
total_words = len(my_tokenizer.word_index) + 1
```

```python
my_tokenizer.word_index
```

### Sequence
- one line consider as a sequence

```python
my_input_sequence = []
#x=1

# Let text file content into lines jher we used \n for seperateion because when we read a text file '\n' added at line seprations
# let iterate through the all lines

for line in alices_dventures_txt.split("\n"):
   
    print(f"\nLine {x}: ",line)
    #x=x+1
    # For each line adding tokenizing then result is numbered list for each waord in line
    # numbers/values assigned using above word_index
    token_list = my_tokenizer.texts_to_sequences([line])[0]
    #print(token_list)

    # now iterate through the each value in the current token_list
    for i in range(1, len(token_list)):
        # genrating cumulative adition lists 10 elts > 9 lists
        my_n_gram_sequence = token_list[:i+1]
        #print(my_n_gram_sequence)

        # list of all n_greams
        my_input_sequence.append(my_n_gram_sequence)
    # print(my_input_sequence)
    # if x==5:
    #     break
```

```python
len(my_input_sequence)
```

```python
for i in range(10) : print(my_input_sequence[i])
```

```python
# Finding len of the longest line
max_sequence_len = max([len(seq) for seq in my_input_sequence])

# pad_sequence = every sequence is padded to length max_sequence_len, padding='pre' = prefix padding as 0s
input_sequences = np.array(pad_sequences(my_input_sequence, maxlen=max_sequence_len, padding='pre'))
```
